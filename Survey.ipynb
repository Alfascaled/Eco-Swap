```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("/content/Kuesioner Pengguna_ EcoSwap App  (Jawaban).xlsx")
```

```python
kolom_skip_manual = ["Timestamp", "Email Address", "Nama :", "Umur","Saran anda untuk Aplikasi Ecoswap ini"]
kolom_skip_auto = [col for col in df.columns if df[col].nunique() <= 2]
kolom_skip = set(kolom_skip_manual + kolom_skip_auto)

kolom_baru = []
q_counter = 1
for col in df.columns:
    if col in kolom_skip:
        kolom_baru.append(col)
    else:
        kolom_baru.append(f"Q{q_counter}")
        q_counter += 1

# Apply nama baru
df.columns = kolom_baru

# Cek hasil
print(df.head())
```

```python
df.head()
```

```python
kolom_q = [col for col in df.columns if col.startswith("Q") and pd.api.types.is_numeric_dtype(df[col])]

# Sekarang aman buat dihitung
rata_per_pertanyaan = df[kolom_q].mean()
```

```python
print("Rata-rata per pertanyaan:")
print(rata_per_pertanyaan)
```

```python
# Paksa konversi semua kolom Q ke numerik
for col in df.columns:
    if col.startswith("Q"):
        df[col] = pd.to_numeric(df[col], errors="coerce")

# Hitung rata-rata per responden (baris)
kolom_q = [col for col in df.columns if col.startswith("Q")]
df["Rata_Rata_Responden"] = df[kolom_q].mean(axis=1)
```
